```python
import pandas as pd
df = pd.read_excel(r"C:\Users\El economista\Documents\GitHub\-BigDataUBA-Grupo-25\TPs\TP 2\df_completo_parte3.xlsx")
```

```python
df["p21"] = df["p21"].astype(float)  # Convertir a float antes de modificar

# Aplicar el coeficiente de ajuste
df.loc[df["ano4"] == 2004, "p21"] *= 314.621568

# Calcular el salario semanal
df["salario_semanal"] = df["p21"] / 40
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filtrar datos válidos y eliminar valores extremos
df_filtrado = df[(df["salario_semanal"].notna()) & 
                 (df["salario_semanal"] > 0) & 
                 (df["salario_semanal"] < df["salario_semanal"].quantile(0.99))]

# Crear figura con dos paneles
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Panel A: Histograma de salario semanal
sns.histplot(df_filtrado["salario_semanal"], bins=30, kde=True, ax=axes[0], color="blue")
axes[0].set_title("Panel A: Histograma del Salario Semanal")
axes[0].set_xlabel("Salario Semanal")
axes[0].set_ylabel("Frecuencia")

# Ajustar formato de números en el eje X para evitar notación científica
axes[0].ticklabel_format(style='plain', axis='x')

# Verificar si hay suficientes valores únicos para KDE
if len(df_filtrado["salario_semanal"].unique()) > 1:
    # Panel B: Distribución Kernel del salario semanal
    sns.kdeplot(df_filtrado["salario_semanal"], fill=True, color="green", ax=axes[1])
    axes[1].set_title("Panel B: Distribución Kernel del Salario Semanal")
    axes[1].set_xlabel("Salario Semanal")
    
    # Ajustar formato de números y límites del eje X
    axes[1].ticklabel_format(style='plain', axis='x')
    axes[1].set_xlim(df_filtrado["salario_semanal"].quantile(0.01), df_filtrado["salario_semanal"].quantile(0.99))
else:
    # Mensaje si no hay suficiente variabilidad
    axes[1].text(0.5, 0.5, "No hay suficiente variabilidad para KDE",
                 ha='center', va='center', fontsize=12)
    axes[1].set_title("Panel B: Distribución Kernel no disponible")
    axes[1].axis("off")  # Oculta los ejes si no hay datos suficientes

plt.tight_layout()
plt.show()
```

```python
# Filtrar datos por año 2004 y 2024
df_2004 = df[df["ano4"] == 2004]
df_2024 = df[df["ano4"] == 2024]

# Función para calcular cantidades con los nuevos valores de Estado
def calcular_cantidades(df):
    return {
        "Cantidad observaciones": len(df),
        "Observaciones con NaN en 'Estado'": df["estado"].isna().sum(),
        "Cantidad de Ocupados (Estado == 1)": len(df[df["estado"] == 1]),
        "Cantidad de Desocupados (Estado == 2)": len(df[df["estado"] == 2]),
        "Cantidad de variables finales": len(df.columns)
    }

# Calcular cantidades para cada año y en total
datos_2004 = calcular_cantidades(df_2004)
datos_2024 = calcular_cantidades(df_2024)
datos_totales = calcular_cantidades(df)

# Crear la tabla con pandas
import pandas as pd
tabla_resumen = pd.DataFrame([datos_2004, datos_2024, datos_totales], index=["2004", "2024", "Total"])

# Mostrar la tabla
print("Tabla 1. Resumen de la base final para la región YYY:")
print(tabla_resumen)
```
